```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://www.amazon.in/s?k=laptop&crid=1YL3M5YH4DUVH&sprefix=laptop%2Caps%2C366&ref=nb_sb_noss_2"  
```

```python
headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}
```

```python
data=[]
```

```python
response=requests.get(url,headers=headers)
response.status_code
```

```python
import time


for page in range(1,3):
    params = {'k': 'laptop', 'page': page}
    response = requests.get(url, headers=headers, params=params)
    soup = BeautifulSoup(response.content, 'html.parser')

    products= soup.find_all('div', {'data-component-type': 's-search-result'})

    # extract the title
    for product in products:
        title_tag = product.find("h2")

        if not title_tag:
           continue
        title_text = title_tag.get_text(strip=True)

        price_tag = product.find("span", {"class": "a-price-whole"})
        price_text = price_tag.get_text(strip=True) if price_tag else "N/A"

        data.append({
            "Title":title_text,
            "Price":price_text
        })

    print(f"Page {page} scraped")
    time.sleep(1)
```

```python
# extracting the brand form the title
for product in data:
    title = product["Title"]
    # Simple brand extraction assuming brand is the first word in the title
    brand = title.split()[0] if title else "Unknown"
    product["Brand"] = brand
    # extract the ram from the title
for product in data:
    title = product["Title"]
    ram_match = re.search(r'(\d+)\s*GB\s*(RAM|DDR\d+|LPDDR\d+)', title)
    ram = ram_match.group(1) if ram_match else "N/A"
    product["RAM"] = ram
for product in data:
    print("Title: ",product["Title"])
    print("Price: ",product["Price"])
    print("Brand: ",product["Brand"])
    print("RAM: ",product["RAM"])
    print("-"*80)
```

```python
# EXtracting Rating improved
for product in products:
    rating_tag = product.find("span", {"class": "a-icon-alt"})
    rating_text = rating_tag.get_text(strip=True) if rating_tag else "N/A"
    print("Rating: ",rating_text)
    print("-"*80)
```

```python
#extract storage from the title
for product in data:
    title = product["Title"]
    storage_match = re.search(r'(\d+)\s*GB\s*(SSD|HDD)', title)
    storage = storage_match.group(1) if storage_match else "N/A"
    product["Storage"] = storage
for product in data:
    print("Storage: ",product["Storage"])
    print("-"*80)
```

```python
#extract color from the title
for product in data:
    title = product["Title"]
    color_match = re.search(r'\b(Black|Silver|Gray|White|Blue|Red|Green|Yellow|Pink)\b', title, re.IGNORECASE)
    color = color_match.group(1) if color_match else "N/A"
    product["Color"] = color
for product in data:
    print("Color: ",product["Color"])
    print("-"*80)
```

```python
#extract windows veraion from the title improved version
for product in data:
    title = product["Title"]
    windows_match = re.search(r'\b(Windows\s*\d{2}|Windows\s*11|Windows\s*10|Windows\s*8|Windows\s*7)\b', title, re.IGNORECASE)
    windows_version = windows_match.group(1) if windows_match else "N/A"
    product["Windows_Version"] = windows_version
    print("Windows Version: ",product["Windows_Version"])
    print("-"*80)
```
